####**FIND MEAN,VARIANCE AND STANDARD DEVIATION OF TEMPERATURE(C)**

```python
#mean
mean_temp=np.mean(data['Temperature (C)'])
print(f"mean temperature is: {mean_temp}")

#variance
var_temp=np.var(data['Temperature (C)'])
print(f"variance temperature is: {var_temp}")

#Standad deviation
std_temp=np.std(data['Temperature (C)'])
print(f"standard deviation temperature is: {std_temp}")
```

####**FIND MEAN,VARIANCE AND STANDARD DEVIATION OF TEMPERATURE(C) FOR EVERY MONTH**

```python
for i in range(1,13):
  month=data.loc[data['month']==i]['Temperature (C)']
  print(f"for month {str(i)}")
  print(f"the mean temp is {str(np.mean(month))}")
  print(f"the variance temp is {str(np.var(month))}")
  print(f"the standard deviation temp is {str(np.std(month))}")
```
