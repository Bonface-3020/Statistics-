```python
data=pd.read_csv('Titanic Dataset.csv')
data.head(10)
```

```python
data.info()
```

```python
minimum_age=data['Age'].min()
print(minimum_age)
maximum_age=data['Age'].max()
print(maximum_age)
```

```python
bin=[0,15,30,45,60,75]
data['binned_age']=pd.cut(data['Age'],bins=bin)
data.head()
```

```python
age_labels=['young','young-Adults','Midde Aged','Middle Older-Age','senior']
data['binned_age']=pd.cut(data['Age'],bins=bin,labels=age_labels,right=False)
data.head()
```

```python
data['binned_age'].value_counts().plot(kind='bar')
plt.title('Age Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
